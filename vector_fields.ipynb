# Code for Visualizing Vector Fields

Importing numpy and matplotlib.

```python
################################################################

import numpy as np
import matplotlib.pyplot as plt

################################################################
```

## Define your variables. Defaults are currently in place.

Read comments for some notes on formatting. 

```python
################################################################

# choose plot title
title = 'Vector Field Computer Activity'

# choose grid size (N x M) keep square and even for best results
userN = 20
userM = 20

#choose number of charges
num = 3

# choose charges: 1 for outward arrows, -1 for inward arrows
# write as q = [q1, q2, .. , qN]
q = [1, -1, 1]

# choose coordinates of charges
# write as x = [x1, x2, .. , xN] and y = [y1, y2, .. , yN]
x = [4, 2, 5]
y = [4, 3, 6]

################################################################
```

## This code creates the coordinate plane, charge points, and field arrows.

```python
################################################################

N = int(userN / 2)
M = int(userM / 2)

# setting up coordinates
X = np.arange(0, N, 1)
Y = np.arange(0, M, 1)
X, Y = np.meshgrid(X, Y)

# strength of the field 
Ex = np.zeros((N, M))
Ey = np.zeros((N, M))

# computing
Q = [[], []]  # to store charges coordinates in loop

for n in range(0, num): 
    
    qq = q[n]
    qx, qy = x[n], y[n]
    
    print(qq, qx, qy)
    
    Q[0].append(qy)
    Q[1].append(qx)
    
    for i in range(N):
        for j in range(M):
            denom = ((i - qx) ** 2 + (j - qy) ** 2) ** 1.5
            if denom != 0: 
                Ex[i, j] += qq * (j - qy) / denom
                Ey[i, j] += qq * (i - qx) / denom
                
################################################################
```

## This last bit colors and normalizes the field, and prints the image.

Picture should appear below.

```python
################################################################

# making arrows different colors, color determined by magnitude
C = np.hypot(Ex, Ey)

# normalizing: making sure arrows are the same length
E = (Ex ** 2 + Ey ** 2) ** .5
Ex = Ex / E
Ey = Ey / E

# drawing the plot
plt.figure(figsize = (userN/2, userM/2))
plt.plot(*Q, 'ko') # draws point charges
plt.quiver(X, Y, Ex, Ey, C) # draws arrows
plt.title(title)
plt.axis('equal')
plt.axis('on')
plt.show()

################################################################
```
