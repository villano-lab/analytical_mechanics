```python
#we may need some code in the ../python directory and/or matplotlib styles
import sys
sys.path.append('../python/')

#matplotlib for plotting
import matplotlib as mpl
from matplotlib import pyplot as plt
plt.style.use('mplstyles/stylelib/standard.mplstyle')

#other computational libraries
import numpy as np
```

```python
R=1
omega=0.1

def x(t,o,r):
    return r*(1-np.cos(o*t))

def y(t,o,r):
    return r*o*t - r*np.sin(o*t)

xro = lambda t: x(t,omega,R)
yro = lambda t: y(t,omega,R)

print(xro(1))

vecxro = np.vectorize(xro)
vecyro = np.vectorize(yro)

print(vecxro([1.0,2.0]))
```

```python
#set up a 1d plot
fig,axes = plt.subplots(1,1,figsize=(9.0,8.0),sharex=True)
ax1 = axes


T=np.arange(0,1000,0.01)


ax1.plot(vecxro(T),vecyro(T),color='orange',linestyle="-",linewidth=2, \
         label='primed frame')









ax1.set_yscale('linear')
#ax1.set_yscale('log')
ax1.set_xlim(-10*R, 20*R) 
ax1.set_ylim(-10*R,100*R)
ax1.set_xlabel(r'x-Distance [m]')
ax1.set_ylabel('y-Distance [m]')
ax1.grid(True)
ax1.yaxis.grid(True,which='minor',linestyle='--')
ax1.legend(loc=1,prop={'size':12})
#ax1.legend(bbox_to_anchor=(1.04,1),borderaxespad=0,prop={'size':22})

for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.tight_layout()
#plt.savefig('figures/figure.png')
plt.show()
```
